```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV files for sparse and dense graphs
sparse_data = pd.read_csv('sparse_graph_results.csv', names=['Vertices', 'Edges', 'Runtime'])
dense_data = pd.read_csv('dense_graph_results.csv', names=['Vertices', 'Edges', 'Runtime'])

# Theoretical runtime (O(V^2)), scaling down to fit on the plot with empirical data
def theoretical_runtime(V, scale_factor=1e-9):
    return scale_factor * V**2

def identify_outliers(data, threshold_factor=1.6):
    # Compute the log of the data
    log_data = np.log10(data)
    mean = np.mean(log_data)
    std = np.std(log_data)
    z_scores = (log_data - mean) / std
    return np.abs(z_scores) > threshold_factor

# Identify outliers in sparse and dense graph runtimes
outliers_sparse = identify_outliers(sparse_data['Runtime'], threshold_factor=1.6)
outliers_dense = identify_outliers(dense_data['Runtime'], threshold_factor=1.6)


# Plot empirical vs theoretical runtime for sparse graphs
plt.figure(figsize=(12, 6))

# Sparse Graphs
plt.subplot(1, 2, 1)
plt.plot(sparse_data['Vertices'][~outliers_sparse], sparse_data['Runtime'][~outliers_sparse], label='Empirical (Sparse Graph)', marker='o', color='b')
plt.plot(sparse_data['Vertices'][outliers_sparse], sparse_data['Runtime'][outliers_sparse], label='Outliers', marker='x', linestyle='None', color='r')
plt.plot(sparse_data['Vertices'], theoretical_runtime(sparse_data['Vertices']), label='Scaled Theoretical O(V^2)', linestyle='--', color='r')
plt.title('Runtime vs Vertices (Sparse Graph)')
plt.xlabel('Number of Vertices')
plt.ylabel('Runtime (seconds)')
plt.ylim(0.00, 0.02)  # Set y-axis limit for sparse graph
plt.legend()

# Dense Graphs
plt.subplot(1, 2, 2)
plt.plot(dense_data['Vertices'][~outliers_dense], dense_data['Runtime'][~outliers_dense], label='Empirical (Dense Graph)', marker='o', color='g')
plt.plot(dense_data['Vertices'][outliers_dense], dense_data['Runtime'][outliers_dense], label='Outliers', marker='x', linestyle='None', color='r')
plt.plot(dense_data['Vertices'], theoretical_runtime(dense_data['Vertices']), label='Scaled Theoretical O(V^2)', linestyle='--', color='r')
plt.title('Runtime vs Vertices (Dense Graph)')
plt.xlabel('Number of Vertices')
plt.ylabel('Runtime (seconds)')
plt.ylim(0.00, 0.025)  # Set y-axis limit for dense graph
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
sparse_data = pd.read_csv('sparse2V_graph_results_heap.csv', names=['Vertices', 'Edges', 'Runtime'])

# Remove outliers based on Z-score or predefined threshold (adjust threshold as needed)
def remove_outliers(data, threshold=1.5):
    z_scores = np.abs((data - np.mean(data)) / np.std(data))
    return z_scores < threshold

# Filter out the outliers from the data
filtered_runtime = sparse_data['Runtime'][remove_outliers(sparse_data['Runtime'], threshold=1.5)]
filtered_vertices = sparse_data['Vertices'][remove_outliers(sparse_data['Runtime'], threshold=1.5)]
filtered_edges = sparse_data['Edges'][remove_outliers(sparse_data['Runtime'], threshold=1.5)]

# Theoretical runtime O((V + E) log V), applying a linear transformation to fit the empirical data
def theoretical_runtime(V, E, scale_factor=1e-2, linear_adjustment=0.3, constant_offset= -1e-4):
    return constant_offset + scale_factor * linear_adjustment * (V + E) * np.log(V)

# Plot empirical vs theoretical runtime for sparse graphs
plt.figure(figsize=(8, 6))

# Plot empirical runtime without outliers
plt.plot(filtered_vertices, filtered_runtime, label='Empirical Time Taken (Sparse 2V)', marker='o', color='b')

# Plot theoretical runtime with new scale factor and offset
plt.plot(filtered_vertices, theoretical_runtime(filtered_vertices, filtered_edges, scale_factor=1e-7, linear_adjustment=1.6), 
         label='Theoretical Time Complexity (Sparse)', linestyle='--', color='r')

# Set y-axis to log scale to visualize better
plt.yscale('log')

plt.title('Empirical vs Theoretical Time Complexity for Sparse Graph (2V Edges)')
plt.xlabel('Number of Vertices (V)')
plt.ylabel('Time Taken (seconds)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Load the CSV file
sparse_data = pd.read_csv('sparse2V_graph_results_heap.csv', names=['Vertices', 'Edges', 'Runtime'])

# Remove outliers based on Z-score or predefined threshold (adjust threshold as needed)
def remove_outliers(data, threshold=1.5):
    z_scores = np.abs((data - np.mean(data)) / np.std(data))
    return z_scores < threshold

# Filter out the outliers from the data
filtered_runtime = sparse_data['Runtime'][remove_outliers(sparse_data['Runtime'], threshold=1.5)]
filtered_vertices = sparse_data['Vertices'][remove_outliers(sparse_data['Runtime'], threshold=1.5)]
filtered_edges = sparse_data['Edges'][remove_outliers(sparse_data['Runtime'], threshold=1.5)]

# Create the feature for (V + E) * log(V)
V_plus_E_log_V = (filtered_vertices + filtered_edges) * np.log(filtered_vertices)

# Reshape the data for fitting
V_plus_E_log_V = V_plus_E_log_V.values.reshape(-1, 1)
runtime = filtered_runtime.values.reshape(-1, 1)

# Fit a linear model
model = LinearRegression()
model.fit(V_plus_E_log_V, runtime)

# Predict the best-fit line
best_fit_runtime = model.predict(V_plus_E_log_V)

# Plot the results
plt.figure(figsize=(8, 6))

# Plot empirical runtime without outliers
plt.plot(filtered_vertices, filtered_runtime, label='Empirical Time Taken (Sparse 2V)', marker='o', color='b')

# Plot best-fit line for the empirical data
plt.plot(filtered_vertices, best_fit_runtime, label='Best Fit Line (O((V + E) log V))', linestyle='--', color='g')

# Set y-axis to log scale to visualize better
plt.yscale('log')

plt.title('Empirical vs Best-Fit Time Complexity for Sparse Graph (2V Edges)')
plt.xlabel('Number of Vertices (V)')
plt.ylabel('Time Taken (seconds)')
plt.legend()

plt.tight_layout()
plt.show()
```
